# **Mini ChatGPT with Static Long Dataset**

```
import difflib

print("="*50)
print("          MINI CHATGPT (STATIC VERSION)")
print("="*50)
print("Ask questions related to AI, ML, Python, Data Science.")
print("Type 'exit' to stop.\n")

# Long Static Dataset
dataset = {
    "what is artificial intelligence":
        "Artificial Intelligence (AI) is the simulation of human intelligence in machines that are programmed to think and learn.",

    "what is machine learning":
        "Machine Learning is a subset of AI that allows systems to learn and improve automatically from experience without being explicitly programmed.",

    "what is deep learning":
        "Deep Learning is a specialized subset of machine learning that uses neural networks with many layers.",

    "what is python":
        "Python is a high-level programming language used for web development, AI, automation, and data science.",

    "what is data science":
        "Data Science is a field that uses scientific methods, algorithms, and systems to extract knowledge from structured and unstructured data.",

    "what is nlp":
        "Natural Language Processing (NLP) is a field of AI that enables computers to understand human language."
}
def find_best_match(user_question):
    questions = dataset.keys()
    match = difflib.get_close_matches(user_question, questions, n=1, cutoff=0.5)
    if match:
        return dataset[match[0]]
    else:
        return "Sorry, I don't have information about that in my dataset."
# Chat Loop
while True:
    user_input = input("You: ").lower()

    if user_input == "exit":
        print("Bot: Chat session ended.")
        break

    answer = find_best_match(user_input)
    print("Bot:", answer)
```
